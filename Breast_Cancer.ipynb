```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# Compute the correlation matrix
correlation_matrix = df.corr()

# Extract correlations with 'diagnosis'
correlation_with_diagnosis = correlation_matrix['diagnosis'].sort_values(ascending=False)

# Plot the correlation values with 'diagnosis'
plt.figure(figsize=(12, 8))
sns.barplot(x=correlation_with_diagnosis.index, y=correlation_with_diagnosis.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Correlation of Features with Diagnosis')
plt.ylabel('Correlation Coefficient')
plt.xlabel('Features')
plt.show()
```

```python
# Select features with high correlation with 'diagnosis'
high_correlation_features = correlation_with_diagnosis[abs(correlation_with_diagnosis) > 0.5]
print("Highly correlated features:\n", high_correlation_features)
```

```python
top_4_features = [
    'concave points_worst',
    'perimeter_worst',
    'concave points_mean',
    'radius_worst'
]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes = axes.flatten()

for i, feature in enumerate(top_4_features):
    sns.boxplot(x='diagnosis', y=feature, data=df, palette='coolwarm', ax=axes[i])
    axes[i].set_title(f'Box plot of {feature} by Diagnosis')
    axes[i].set_xlabel('Diagnosis')
    axes[i].set_ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers for each of the top 4 features
for feature in top_4_features:
    df = remove_outliers(df, feature)

# Verify the data after removing outliers
print(df.describe())
```

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot box plots for the top 4 features after removing outliers
for i, feature in enumerate(top_4_features):
    sns.boxplot(x='diagnosis', y=feature, data=df, palette='coolwarm', ax=axes[i])
    axes[i].set_title(f'Box plot of {feature} by Diagnosis (Outliers Removed)')
    axes[i].set_xlabel('Diagnosis')
    axes[i].set_ylabel(feature)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
x = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score, roc_auc_score, roc_curve
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Recall:", recall)
print("Precision:", precision)
print("F1 Score:", f1)
```
